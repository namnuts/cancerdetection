```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Optional, for enhanced visualizations
```

```python
plt.style.use('ggplot')  # You can choose from styles like 'seaborn', 'classic', etc.
```

```python
# Checking upon diagnosis
diagnosis_counts = df['diagnosis'].value_counts()

# Create a bar plot
plt.figure(figsize=(6,4))
diagnosis_counts.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keeps the labels horizontal
plt.show()
```

```python
# Create a pie chart
plt.figure(figsize=(6,6))
diagnosis_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=140)
plt.title('Diagnosis Distribution')
plt.ylabel('')  # Removes the y-label
plt.show()
```

```python
# Separate the data based on diagnosis
malignant = df[df['diagnosis'] == 'M']
benign = df[df['diagnosis'] == 'B']

# Plot histograms
plt.figure(figsize=(8,6))
plt.hist(malignant['radius_mean'], bins=10, alpha=0.7, label='Malignant', color='salmon')
plt.hist(benign['radius_mean'], bins=10, alpha=0.7, label='Benign', color='skyblue')
plt.title('Distribution of Radius Mean by Diagnosis')
plt.xlabel('Radius Mean')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(6,4))
sns.boxplot(x='diagnosis', y='area_mean', data=df, palette=['salmon', 'skyblue'])
plt.title('Box Plot of Area Mean by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Area Mean')
plt.show()
```

```python
# Prepare data
data_to_plot = [malignant['area_mean'], benign['area_mean']]

# Create box plot
plt.figure(figsize=(6,4))
plt.boxplot(data_to_plot, labels=['Malignant', 'Benign'], patch_artist=True,
            boxprops=dict(facecolor='salmon', color='salmon'),
            medianprops=dict(color='yellow'))
plt.title('Box Plot of Area Mean by Diagnosis')
plt.ylabel('Area Mean')
plt.show()
```

```python
# Create scatter plot
plt.figure(figsize=(8,6))
plt.scatter(malignant['radius_mean'], malignant['texture_mean'], label='Malignant', color='salmon', alpha=0.7)
plt.scatter(benign['radius_mean'], benign['texture_mean'], label='Benign', color='skyblue', alpha=0.7)
plt.title('Radius Mean vs. Texture Mean by Diagnosis')
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.legend()
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df.drop(['id', 'diagnosis', 'diagnosis_encoded'], axis=1).corr()

# Create a heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
# Select a subset of features
subset = df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'diagnosis']]

# Create a pair plot
sns.pairplot(subset, hue='diagnosis', palette=['salmon', 'skyblue'])
plt.suptitle('Pair Plot of Selected Features by Diagnosis', y=1.02)
plt.show()
```
